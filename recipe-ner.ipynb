```python
import pandas as pd
import spacy
from spacy.tokens import DocBin
import json
import ast

# Load dataset
df = pd.read_csv("recipe-100.csv")  # Replace with your actual file name

# Initialize a blank spaCy model
nlp = spacy.blank("en")
doc_bin = DocBin()

# Process each row and convert to SpaCy format
for _, row in df.iterrows():
    recipe_text = row['ingredients']  # Assuming 'ingredients' contains the text
    ner_labels = ast.literal_eval(row['NER'])  # Convert string list to Python list
    
    entities = []
    for ingredient in ner_labels:
        start_idx = recipe_text.lower().find(ingredient.lower())
        if start_idx != -1:
            end_idx = start_idx + len(ingredient)
            entities.append((start_idx, end_idx, "INGREDIENT"))  # Label all as 'INGREDIENT'
    
    # Create the SpaCy document
    doc = nlp.make_doc(recipe_text)
    ents = [doc.char_span(start, end, label) for start, end, label in entities if doc.char_span(start, end, label)]
    doc.ents = ents
    
    # Add the doc to the training data
    doc_bin.add(doc)

# Save processed data to SpaCy binary format
doc_bin.to_disk("ner_training_data.spacy")
```
